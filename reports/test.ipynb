```python
import pandas as pd
import matplotlib.pyplot as plt

# c d liu t tp CSV
file_path = r'D:\data_analysis\speech_emotion_recognition\src\feature_extraction\merged_train_feature_data.csv'
data = pd.read_csv(file_path)

# To mt t in  nh x cc nhn cm xc thnh mu sc tng ng
emotion_to_color = {'Happy': 'red', 'Sad': 'blue', 'Angry': 'green', 'Neutral': 'yellow', 'Fearful': 'pink', 'Disgusted': 'orange'}

# Lc cc ct c cha 'zcr' trong tn ct
zcr_features = [col for col in data.columns if 'zcr' in col]

# V biu  phn b cho tng tnh nng 'zcr'
for feature in zcr_features:
    plt.figure(figsize=(10, 6))
    for emotion, color in emotion_to_color.items():
        # Lc d liu cho mi nhn cm xc
        subset = data[data['label'] == emotion]
        # V biu  phn b cho tng nhn cm xc
        plt.hist(subset[feature], bins=30, alpha=0.7, color=color, label=emotion)
    
    # Thm tiu  v ch thch
    plt.title(f'Distribution of {feature} across different emotions')
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.legend()
    # Hin th biu 
    plt.show()
```
