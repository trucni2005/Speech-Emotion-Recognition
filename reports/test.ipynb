```python
import matplotlib.pyplot as plt
import librosa
import librosa.display
import numpy as np
```

```python
def plot_audio(file_path):
    # Đọc file âm thanh và tần số lấy mẫu
    y, sr = librosa.load(file_path, sr=None)

    # Biểu đồ âm thanh theo thời gian
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    librosa.display.waveshow(y, sr=sr)
    plt.title('Waveform of Audio File')
    plt.xlabel('Time (s)')
    plt.ylabel('Amplitude')

    # Biểu đồ phổ âm thanh
    plt.subplot(2, 1, 2)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Spectrogram of Audio File')
    plt.xlabel('Time (s)')
    plt.ylabel('Frequency (Hz)')

    plt.tight_layout()
    plt.show()
```

```python
file_path = r"D:\data_analysis\speech_emotion_recognition\data\EnglishDataset\train_test_splited_data\augmented\train\03-01-01-01-01-01-01.wav"
plot_audio(file_path)
```
