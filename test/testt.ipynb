```python
import os
import matplotlib.pyplot as plt

def count_wav_files(folder):
    # Đếm số lượng file .wav trong một folder
    wav_count = sum(1 for file in os.listdir(folder) if file.endswith('.wav'))
    return wav_count

def process_folder(input_folder):
    # Xử lý mỗi folder: đọc số lượng file .wav và vẽ biểu đồ
    wav_count = count_wav_files(input_folder)

    # Vẽ biểu đồ cột
    plt.bar(os.path.basename(input_folder), wav_count)

    # Đặt tiêu đề cho biểu đồ
    plt.xlabel('Cảm xúc')
    plt.ylabel('Số lượng mẫu dữ liệu')
    plt.title('Phân bố cảm xúc của dữ liệu sau khi làm sạch và chuẩn hóa')

    # Hiển thị giá trị trên mỗi cột
    plt.text(os.path.basename(input_folder), wav_count, str(wav_count), ha='center', va='bottom')

def process_folders(input_folder):
    for root, dirs, files in os.walk(input_folder):
        for folder in dirs:
            subdirectory_input = os.path.join(root, folder)

            process_folder(subdirectory_input)
    plt.tight_layout()       # Đảm bảo không bị cắt bớt nội dung
    plt.show()

process_folders(r'D:\data_analysis\speech_emotion_recognition\data\EnglishDataset\normalized_data')
```
